```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def generate_data():
    wilayas = ['Algiers', 'Blida', 'Oran', 'Constantine', 'Annaba']
    quality_grades = ['Commercial', 'Standard', 'Premium']
    products = ['Potatoes', 'Tomatoes', 'Corn', 'Wheat', 'Carrots']
    
    # Generate 10000 listings
    listings = pd.DataFrame({
        'listing_id': [f'L{i:03d}' for i in range(20000)],
        'product_name': np.random.choice(products, 20000),
        'wilaya': np.random.choice(wilayas, 20000),
        'available_quantity_kg': np.random.randint(50, 20000, 20000),
        'quality_grade': np.random.choice(quality_grades, 20000),
        'price_dzd': np.random.randint(300, 1000, 20000),
        'seller_rating': np.random.uniform(3.0, 5.0, 20000),
        'days_listed': np.random.randint(1, 60, 20000)
    })
    
    # Calculate price per kg
    listings['price_per_kg'] = listings['price_dzd'] / listings['available_quantity_kg']
    
    # Save
    listings.to_csv('mock_listings.csv', index=False)
    print("✅ Mock data generated!")

if __name__ == '__main__':
    generate_data()
```
